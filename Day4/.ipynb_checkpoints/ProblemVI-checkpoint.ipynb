*Hint* – you can use [Cramer's V](https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V) to calculate the correlation between categorical variables. `scipy` provides the function [`association`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.contingency.association.html) to calculate the value of Cramer's V. 

*Bonus hint* – Cramer's V is but one tool in your tool box. Can you think of others? Can you build a visualization to help develop intuition to answer this question? 

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, pointbiserialr, spearmanr
from sklearn.preprocessing import LabelEncoder

# Assuming you have redcard_df already loaded
# Define a function to calculate Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

# Calculate Cramér's V for each column against 'Red Cards'
columns = ['Player', 'Club', 'POS', 'RACE']
results = {}

for col in columns:
    results[col] = cramers_v(redcard_df[col], redcard_df['Red Cards'])

# Display Cramér's V results
print("Cramér's V correlations with 'Red Cards':")
print(results)
```

```python
import pandas as pd
import numpy as np
import corner
from sklearn.preprocessing import LabelEncoder

# Assuming redcard_df is already loaded
# Encode categorical columns to numerical values
le = LabelEncoder()
encoded_df = redcard_df.copy()

columns = ['Player', 'Club', 'POS', 'RACE']
for col in columns:
    encoded_df[col + '_encoded'] = le.fit_transform(encoded_df[col])

# Selecting relevant columns for the plot (encoded and 'Red Cards')
plot_data = encoded_df[['Red Cards'] + [col + '_encoded' for col in columns]].values

# Create the corner plot
corner.corner(plot_data, labels=['Red Cards'] + columns, show_titles=True)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Assuming redcard_df is already loaded
# Step 1: Create a combined categorical column
redcard_df['Club_POS'] = redcard_df['Club'] + '_' + redcard_df['POS']
redcard_df['Club_RACE'] = redcard_df['Club'] + '_' + redcard_df['RACE']
redcard_df['POS_RACE'] = redcard_df['POS'] + '_' + redcard_df['RACE']

print(redcard_df)
```

```python
# Step 2: Define a function to calculate Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    print(confusion_matrix)
    print('')
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

# Initialize a dictionary to store the results
results = {}

# Calculate Cramér's V for each column against 'Red Cards'
for col in redcard_df.columns:
    if col != 'Red Cards':  # Skip the 'Red Cards' column itself
        print(col)
        results[col] = cramers_v(redcard_df[col], redcard_df['Red Cards'])

# Sort results from greatest to least
sorted_results = sorted(results.items(), key=lambda item: item[1], reverse=True)

# Display the results one by one
print("Cramér's V correlations with 'Red Cards' (sorted from greatest to least):")
for col, val in sorted_results:
    print(f"{col}: {val:.4f}")

```
